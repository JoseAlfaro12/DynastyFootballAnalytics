```python
import pandas as pd
import numpy as np
```

```python
# Managers and weeks
managers = ["Jose", "Daryl", "Gio", "Brigido", "Luis", "Lalo", "Devonte", "Marvin", "Jacky", "Bryan"]
weeks = list(range(1, 14))  # Weeks 1â€“13

# Create a multi-index dataframe template
rows = []
for owner in managers:
    for week in weeks:
        rows.append({
            "owner": owner,
            "week": week,
            "injured_players": []  # Fill this list with player IDs or names
        })

injury_df = pd.DataFrame(rows)

# Optional: add a placeholder column for number of injured players
injury_df['num_injured'] = injury_df['injured_players'].apply(len)

# Preview
print(injury_df.head(15))
```
