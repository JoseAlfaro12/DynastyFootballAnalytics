```python
week_starts = [
    (1,  '2023-09-07'),
    (2,  '2023-09-14'),
    (3,  '2023-09-21'),
    (4,  '2023-09-28'),
    (5,  '2023-10-05'),
    (6,  '2023-10-12'),
    (7,  '2023-10-19'),
    (8,  '2023-10-26'),
    (9,  '2023-11-02'),
    (10, '2023-11-09'),
    (11, '2023-11-16'),
    (12, '2023-11-23'),
    (13, '2023-11-30'),
    (14, '2023-12-07')
]

week_starts = [(wk, pd.to_datetime(dt)) for wk, dt in week_starts]
```

```python
week_starts
```

```python
def classify_week(trade_date):
    trade_date = pd.to_datetime(trade_date)

    # Pre-season: before first regular-season game
    if trade_date < week_starts[0][1]:
        return 'pre-season'

    # In-season: assign to the *latest* week whose start date has passed
    for i in range(len(week_starts) - 1):
        current_week, current_start = week_starts[i]
        next_week, next_start = week_starts[i + 1]

        if current_start <= trade_date < next_start:
            return f'{current_week + 1}' if trade_date > current_start else f'{current_week}'

    # After Week 14 start
    if trade_date >= week_starts[-1][1]:
        # Trades after Week 14 games end  offseason
        if trade_date > pd.to_datetime('2023-12-11'):
            return 'offseason'
        return 'Week 14'
```

```python
trades_2023['week'] = trades_2023['date'].apply(classify_week)
```

```python
# Get all unique owners
all_owners = pd.concat([trades_2023['receiving_owner'], trades_2023['giving_owner']])
owners = all_owners.unique()

# Initialize pick ownership dictionary
starting_picks = {}
for owner in owners:
    starting_picks[owner] = [
        '2024_1','2024_2','2024_3','2024_4',
        '2025_1','2025_2','2025_3','2025_4',
        '2026_1','2026_2','2026_3','2026_4'
    ]
```

```python
starting_picks
```

```python
# Process trades
for _, trade in trades_2023.iterrows():
    # Receiving owner: add received, remove gave
    starting_picks[trade['receiving_owner']].extend(trade['ro_picks_received'])
    for pick in trade['ro_picks_gave']:
        if pick in starting_picks[trade['receiving_owner']]:
            starting_picks[trade['receiving_owner']].remove(pick)
    
    # Giving owner: add received, remove gave
    starting_picks[trade['giving_owner']].extend(trade['go_picks_received'])
    for pick in trade['go_picks_gave']:
        if pick in starting_picks[trade['giving_owner']]:
            starting_picks[trade['giving_owner']].remove(pick)
```

```python
starting_picks
```

```python
# Create a summary DataFrame
final_picks = pd.DataFrame([
    {'owner': owner, 'final_picks': picks, 'num_picks': len(picks)}
    for owner, picks in starting_picks.items()
])
```

```python
final_picks
```

```python
# Function to sort picks
def sort_picks(picks):
    # Filter out any empty strings
    picks = [p for p in picks if p]
    # Sort by season then round
    picks_sorted = sorted(
        picks,
        key=lambda x: (int(x.split('_')[0]), int(x.split('_')[1]))
    )
    return picks_sorted

# Apply to the final_picks column
final_picks['final_picks'] = final_picks['final_picks'].apply(sort_picks)

# Display the result
final_picks[['owner','final_picks', 'num_picks']].sort_values('num_picks', ascending= False)
```

### Combine Datasets (2023-2025)

```python
common_cols = [
    'date',
    'receiving_owner',
    'giving_owner',
    'ro_picks_received',
    'ro_picks_gave',
    'go_picks_received',
    'go_picks_gave'
]

trades_2023 = trades_2023[common_cols]
trades_2024 = trades_2024[common_cols]
trades_2025 = trades_2025[common_cols]
```

```python
trades_2023 = trades_2023.assign(season=2023)
trades_2024 = trades_2024.assign(season=2024)
trades_2025 = trades_2025.assign(season=2025)
```

```python
all_trades = pd.concat(
    [trades_2023, trades_2024, trades_2025],
    ignore_index=True
)
```

```python
all_trades['date'] = pd.to_datetime(all_trades['date'], errors='coerce')
all_trades = all_trades.sort_values('date').reset_index(drop=True)
```
